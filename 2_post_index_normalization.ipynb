```python
import time
import datetime
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome()
driver.get("https://www.pochta.ru/post-index")
```

```python
import pandas as pd
address = pd.read_csv('Downloads/missing_addr (1).csv', header=None)

address = address.groupby(0, as_index=False).first()

address.shape
```

```python
address.tail()
```

```python
%%time

counter = 0
res = pd.DataFrame()

for index, row in address.iterrows():
    
    addr = ''
    
    try:   
        elem = driver.find_element_by_tag_name('input')
        elem.clear()
        elem.send_keys(row[0])
        time.sleep(5)
        elem_sugg = driver.find_element_by_class_name('input__suggest__element')
        addr = elem_sugg.text.split(',')
        list.reverse(addr)
        addr = ', '.join(addr).strip()
        elem_sugg.click()
        time.sleep(6)
    except:
        pass
    if len(addr) > 0:
        try:
            elem_idx = driver.find_element_by_class_name('index-card__postal-code')
            addr += ''+ ', ' + elem_idx.text
        except:
            pass
    
    item = {}    
    item['address'] = row[0]
    item['address_rus'] = addr
    
    addr = ''
    
    counter+=1
    
    res = res.append(pd.DataFrame(item, index=[0]))
    
    print(counter, datetime.datetime.now(), item)
```

```python
res.to_csv('address_pochta_4.csv', index=False)
```
