```python
import requests
import pandas as pd
import json
```

```python
max_city = 1000000
max_pages = 1000
```

```python
atms = pd.DataFrame()
```

```python
last_active_city = 0

for city in range(1, max_city):
    if city - last_active_city > 2048:
        break
    for page in range(0, max_pages):
        offset = page * 100
        url = 'https://alfabank.ru/ext-json/0.2/atm/list?city='+str(city)+'&limit=100&offset='+ str(offset) +'&mode=array'
        r = requests.get(url)
        
        if int(r.status_code) != 200: 
            print(r.status_code)
        else:
            try:
                j = json.loads(r.content)
                count = j['response']['count']
                if count < offset:
                    break                   
                atms_before = atms.shape[0]
                atms = atms.append(pd.DataFrame(j['response']['data']));
                atms.to_csv('alfa-atms.csv', index=False)
                atms_after = atms.shape[0]
                if atms_after > atms_before:
                    last_active_city = city
            except:
                break
        print(city, offset, atms.shape)
```
