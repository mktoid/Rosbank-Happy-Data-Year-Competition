```python
atms['Адрес'] = atms['Индекс'].fillna('').astype(str) + ' ' + atms['Субъект федерации'].fillna('').astype(str) + ' ' +atms['Название субъекта федерации'].fillna('').astype(str)
atms['Адрес'] = atms['Адрес'] + ' ' + atms['Регион'].fillna('').astype(str) + ' ' +atms['Название региона'].fillna('').astype(str)
atms['Адрес'] = atms['Адрес'] + ' ' + atms['Населенный пункт'].fillna('').astype(str) + ' ' +atms['Название населенного пункта'].fillna('').astype(str)
atms['Адрес'] = atms['Адрес'] + ' ' + atms['Район населенного пункта'].fillna('').astype(str) + ' ' +atms['Название района нас. пункта'].fillna('').astype(str)
atms['Адрес'] = atms['Адрес'] + ' ' + atms['Улица'].fillna('').astype(str) + ' ' +atms['Название улицы'].fillna('').astype(str)
atms['Адрес'] = atms['Адрес'] + ' ' + atms['Дом'].fillna('').astype(str) + ' ' +atms['Номер дома'].fillna('').astype(str)
atms['Адрес'] = atms['Адрес'] + ' ' + atms['Строение'].fillna('').astype(str) + ' ' +atms['Номер строения'].fillna('').astype(str)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import datetime
import time
```

```python
import sys

def geodecoder(query):

    url = 'https://geocode-maps.yandex.ru/1.x/?geocode=' + query
    r = requests.get(url)
    
    sleep_delay = 1
    
    while int(r.status_code) != 200: 
        print(r.status_code, sleep_delay)
        time.sleep(sleep_delay)
        if sleep_delay < 60 * 60 :
            sleep_delay = sleep_delay * 2

    content = r.content
    soup = BeautifulSoup(content, "xml")

    lat = None
    lon = None

    try:
      coords = soup.find('pos').text.split(' ')
      lat = coords[1]
      lon = coords[0]
    except:
      pass

    return lon, lat


from datetime import datetime
import numpy as np

atms = pd.read_csv('sber-atms-v2.csv')
atms = atms.sample(frac=1)

for index, row in atms.iterrows():
    if np.isnan(row['lon']):
        atms.at[index,'lon'], atms.at[index,'lat'] = geodecoder(row['Адрес'])
        print(datetime.now(), atms[atms['lat'] > 0].shape[0])
        atms.to_csv('sber-atms-v2.csv', index=False)
```
