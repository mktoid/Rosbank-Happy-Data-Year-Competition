```python
population = pd.read_excel('Tabl-01-18.xls')
population.columns = ['region', 'population2018','urban2018', 'rural2018', 'overal2017','urban2017', 'rural2017']
population = population.dropna()
population.region = population.region.astype(str).apply(lambda x: x.replace('  ',' '))
population.region = population.region.astype(str).apply(lambda x: x.replace('включая авт. округа',''))
population.region = population.region.astype(str).apply(lambda x: x.replace('без авт. округов',''))
population.region = population.region.astype(str).apply(lambda x: x.replace('авт. округ','автономный округ'))
population.region = population.region.astype(str).apply(lambda x: x.replace('без автономный округов',''))
population.region = population.region.astype(str).apply(lambda x: x.replace('г.','').strip())
population.drop(['overal2017','urban2017','rural2017'], axis=1, inplace=True)
population
```

```python
salary = pd.read_excel('t2.xlsx')
salary = salary[['Обновлено 19.11.2018', 'Unnamed: 69']].dropna()
salary.columns = ['region', 'salary']
salary.region = salary.region.astype(str).apply(lambda x: x.replace('г.','').strip())
salary.region = salary.region.astype(str).apply(lambda x: x.replace('авт.округ','автономный округ'))
salary.region = salary.region.astype(str).apply(lambda x: x.replace('  ',' '))
salary
```

```python
# формат записи регионов отличается, поэтому перейдем к тому, как указано в геокодере Яндекса
```

```python
yandex = pd.read_csv('address_combined_pochta_yandex_centers.csv')

regions = set()
regions.add('Российская Федеpация')

for item in yandex.AdministrativeAreaName.dropna().unique():
    regions.add(item)
    
for item in yandex.SuperAdministrativeAreaName.dropna().unique():
    regions.add(item)
    
regions
```

```python
fuzzy_regions = set()
```

```python
for region in population.region.unique():
    if region not in regions:
        fuzzy_regions.add(region)
        print(region)
```

```python
for region in salary.region.unique():
    if region not in regions:
        fuzzy_regions.add(region)
        print(region)
```

```python
%%time
try:
    model
except NameError:
    from gensim.models import FastText
    model = FastText.load_fasttext_format('datasets/cc.ru.300.bin') # pre-trained vectors from https://fasttext.cc/docs/en/crawl-vectors.html
```

```python
fuzzy_regions
```

```python
%%time

df = pd.DataFrame()

for r2 in fuzzy_regions:
    rdf = pd.DataFrame()
    for r1 in regions:
        item = {}
        item['yandex_region'] = r1
        item['region'] = r2
        item['distance'] =  model.wv.distance(str(r1), str(r2))
        
        rdf = rdf.append(pd.DataFrame(item, index=[0]))
    df = df.append(rdf.sort_values('distance').iloc[0])
```

```python
df
```

```python
# местами кривовато из-за того, что регионов нет в выборке по банкоматам, стараемся вытянуть что есть

replacement = df[df['distance'] < .144]
replacement
```

```python
def normalize_region(region):
    if replacement[replacement['region'] == region].shape[0] == 0:
        return region
    else:
        return replacement[replacement['region'] == region]['yandex_region'].values[0]
```

```python
%%time
population.region = population.region.apply(normalize_region)
salary.region = salary.region.apply(normalize_region)
```

```python
population.groupby('region', as_index=False).max().to_csv('population_v3.csv', index=False)
salary.groupby('region', as_index=False).max().to_csv('salary_v3.csv', index=False)
```
