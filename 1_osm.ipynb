```python
import osmread
osm_file = osmread.parse_file('russia-latest.osm.pbf')
```

```python
tagged_nodes = [
    entry
    for entry in osm_file
    if isinstance(entry, osmread.Node)
    if len(entry.tags) > 0
]
```

```python
import pickle

with open('osm_tagged_nodes.pickle', 'wb') as fout:
    pickle.dump(tagged_nodes, fout, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
import pandas as pd
tagged_nodes = pd.read_pickle('osm_tagged_nodes.pickle')
```

```python
tagged_nodes[1024]
```

```python
import numpy as np

osm_tags = []
for i in range(len(tagged_nodes)):
    if (len(tagged_nodes[i].tags)>0):
        for j in tagged_nodes[i].tags:
            osm_tags.append(j)
tags, count = np.unique(osm_tags,return_counts=True)

tags_count = pd.DataFrame({'tag':tags, 
                           'count':count}).sort_values(by='count',ascending=False).reset_index(drop=True)
```
